### Dropping the first 2 columns as its not required

```python
stroke_data.drop(columns=['id','gender'],inplace=True)
```

```python
stroke_data.head()
```

### Checking for columns that is useful or not

```python
stroke_data=pd.get_dummies(stroke_data, columns=['ever_married','work_type','Residence_type', 'smoking_status'])
```

```python
stroke_data
```

```python
stroke_data= stroke_data.dropna()
```

### Scaling the values

```python
X = stroke_data.drop(columns=['stroke'])
y = stroke_data['stroke']

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X,y, test_size =0.2, random_state=2)
```

```python
X
```

```python
X_train.shape, X_test.shape, y_train.shape, y_test.shape
```

```python
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
X_train_scaled
```

```python
import tensorflow
from tensorflow import keras
from tensorflow.keras import Sequential
from tensorflow.keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(3,activation='sigmoid',input_dim =18))
model.add(Dense(1,activation='sigmoid'))
```

```python
model.summary()
```

```python
model.compile(loss='binary_crossentropy', optimizer ='Adam')
```

```python
model.fit(X_train_scaled, y_train, epochs=10)
```

Checking weights and biases

```python
model.layers[1].get_weights()
```

### Prediction

```python
y_log = model.predict(X_test_scaled)
```

```python
y_pred = np.where(y_log>0.5, 1,0)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y_test,y_pred)
```
